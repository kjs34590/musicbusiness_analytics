## 01 Spotify API를 활용한 데이터 수집
### 1) API 개념 이해
**API(Application Programming Interface) :**

두 개의 시스템이 서로 상호작용하기 위한(데이터를 주고받기 위한) 인터페이스

> endpoint : 리소스를 액세스하는 경로/방법
>
> method : 리소스 접근에 허용된 행위 (GET/POST/PUT/DELETE/HEAD)
> - GET : 해당 리소스를 조회하고 정보를 가져온다.
> - HEAD : GET 방식과 동일하나, 응답코드와 HEAD만 가져온다.
> - POST : 요청된 리소스를 생성한다.
> - PUT : 요청된 리소스를 업데이트한다.
> - DELETE : 요청된 리소스를 삭제한다.
>
> parameter : endpoint를 통해 request를 할 때 전달하는 옵션
> - Header : Request header에 포함되는 Parameter로 주로 Authorization과 관련
> - Path : Query String 이전에 Endpoint Path 안에 포함되는 변수 (ex:{id})
> - Query String : Query String 이후 포함되는 파라미터
> - Request Body : Request Body 안에 포함되는 파라미터. 주로 json 형태

### 2) Spotify API 사용 준비
1. Spotify 개발자 사이트(https://developer.spotify.com)에서 API 접근을 위한 ID/PW 발급
2. 가입 > 로그인 > Dashboard에서 Spotify App 생성
3. Client ID와 secret KEY 획득 (SHOW CLIENT SECRET 버튼 클릭)

> 방식 : App에 표시된 clien ID와 secret KEY를 Spotify에게 넘기면, Spotify가 다시 Access Token 반환. 이 Access Token을 통해 Data를 호출한다.
