```python
from selenium import webdriver
from time   import sleep, time
from random import randint
```

```python
# 뮤직카우 데이터 로드
mcow = pd.read_csv('./data/musiccow_list.csv')
len(mcow)
```

```python
# chromedriver 변수 지정
# 검색용 url 변수 지정
path = 'C:/users/kjs34/chromedriver.exe'
driver = webdriver.Chrome(path)
url = 'https://genie.co.kr/search/searchMain?query='
```

```python
# 받을 feature 빈 list 생성
titles = []
albums = []
artists = []
genres = []
publishers = []
agencies = []
pub_dates = []
streams = []
listeners = []

err_title = []
err_artist = []

# 시간 표시
now = datetime.datetime.now() 
formattedDate = now.strftime("%Y%m%d")

# 완료, 에러 수 표시를 위한 변수 지정
done = 0
error = 0

# 시작시간
print('START DATE : ', now)

# 크롤링 반복문
for i in range(0,len(mcow)) :
    driver.get(url+mcow['title'][i]+' '+mcow['artist'][i])
    sleep(1)
    try : 
        btn_basic = driver.find_elements_by_css_selector('td.link > a.btn-basic.btn-info')[0]
        btn_basic.click()
        sleep(1.5)
        songTtl = driver.find_element_by_css_selector('div > h2.name').text
        listen_total = int(driver.find_elements_by_css_selector('div.total > div > p')[0].text.replace(',',''))
        stream_total = int(driver.find_elements_by_css_selector('div.total > div > p')[1].text.replace(',',''))
        albumBtn = driver.find_elements_by_css_selector('li > span.value > a')[1]
        albumBtn.click()
        sleep(1.5)
        albumTtl = driver.find_element_by_css_selector('div.info-zone > h2.name').text
        albumTag = driver.find_elements_by_css_selector('ul.info-data > li >span.value')

        # 데이터 추가
        titles.append(songTtl)
        albums.append(albumTtl)
        artists.append(albumTag[0].text)
        genres.append(albumTag[1].text)
        publishers.append(albumTag[2].text)
        agencies.append(albumTag[3].text)
        pub_dates.append(albumTag[4].text)
        listeners.append(listen_total)
        streams.append(stream_total)
        print('{}번곡 : {}-{} DONE'.format(i, mcow['title'][i], mcow['artist'][i]))
        done = done + 1
        sleep(0.5)
    except : 
        print('{}번곡 : {}-{} ERROR !'.format(i, mcow['title'][i], mcow['artist'][i]))
        err_title.append(mcow['title'][i])
        err_artist.append(mcow['artist'][i])
        error = error + 1
        sleep(0.5)

globals()['musiccow_genie_{}'.format(formattedDate)] = pd.DataFrame({'title' : titles,
                                                               'artist' : artists,
                                                               'album' : albums,
                                                               'genre' : genres,
                                                               'publisher' : publishers,
                                                               'agency' : agencies,
                                                               'pub_date' : pub_dates,
                                                               'listeners' : listeners,
                                                               'streams' : streams})
globals()['error_list'] = pd.DataFrame({'title' : err_title, 'artist' : err_artist})

now = datetime.datetime.now() 
formattedDate = now.strftime("%Y%m%d")
print('[result]','-'*50)
print('musiccow_genie_{} ALL DONE'.format(formattedDate))
print('END DATE : ', now)
print('DONE : ', done)
print('ERROR : ', error)
print('ERROR LIST : \n', error_list)
```
