```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
from selenium.webdriver.common.keys import Keys
from time import time, sleep
```

```python
song_data = pd.read_csv('./data/musicow_song_artist.csv')
```

```python
len(song_data['title'])
```

```python
driver = webdriver.Chrome('C:/users/kjs34/chromedriver.exe')
song_kw = []
artist_kw = []
name_list = []
view_list = []

for i in range(0, len(song_data['title'])-1):
    url = 'https://www.youtube.com/results?search_query={} {}'.format(song_data['title'][i], song_data['artist'][i])
    driver.get(url)
    sleep(0.5)
    
    soup = bs(driver.page_source, 'html.parser')
    name = soup.select('a#video-title')
    view = soup.select('a#video-title')
    
    for j in range(len(name)):
        song_kw.append(song_data['title'][i])
        artist_kw.append(song_data['artist'][i])
        name_list.append(name[j].text.strip())
        view_list.append(view[j].get('aria-label').split()[-1][:-1].replace(',',''))
    print('DONE {} : {}'.format(i, song_data['title'][i]))

youtubeDic = {
    'song': song_kw,
    'artist': artist_kw,
    'yt_title' : name_list,
    'view': view_list
}

youtubeDf = pd.DataFrame(youtubeDic)
youtubeDf
```

```python
youtubeDic = {
    'song': song_kw,
    'artist': artist_kw,
    'yt_title' : name_list,
    'view': view_list
}

youtubeDf = pd.DataFrame(youtubeDic)
youtubeDf
```

```python
# view == '없' (조회수 없음) 값을 0으로 대체
youtubeDf['view']  = np.where(youtubeDf['view'] == '없', '0', youtubeDf['view'])
```

```python
# 대체 여부 확인
youtubeDf[youtubeDf['view'] == '0']['view']
```

```python
youtubeDf = youtubeDf.astype({'view':'float'})
youtubeDf
```

```python
pd.options.display.float_format = '{:.5f}'.format
```

```python
youtube_mean = youtubeDf.groupby(['song', 'artist']).mean('view').sort_values('view', ascending=False)
```

```python
youtube_mean.to_csv('./data/youtube_mean.csv', encoding='cp949')
```
