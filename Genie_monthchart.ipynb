```python
path = 'C:/users/kjs34/chromedriver.exe'
driver = webdriver.Chrome(path)
url = 'https://www.genie.co.kr/chart/top200?ditc=M&rtm=N&ymd='
dates = ['20190101']
```

### 2. Crawling & Scrapping
> **삼중 for 문 활용**
> 1. 연도-월별 페이지 변경 Loop
> 2. 1-50위, 51-100위, 101-150위, 151-200위 버튼 클릭 Loop (*4)
> 3. 페이지 내의 곡별 앨범 클릭 -> 정보 스크랩 -> 뒤로 가기 Loop (*50)

> **최종 데이터 정의**
> - chart : 차트 기간
> - title : 곡 이름
> - rank : 순위
> - album : 앨범
> - artist : 아티스트
> - genre : 장르
> - publisher : 발매사(유통사)
> - agency : 기획사
> - date : 곡 발매일

```python
# album click -> scrap -> 뒤로 가기
for i in range(0,len(dates)) :
    driver.get(url+dates[i])
    chartdate = []
    ranks = []
    titles = []
    albums = []
    artists = []
    genres = []
    publishers = []
    agencies = []
    pub_dates = []
    for j in range(0, 4) :
        driver.find_elements_by_css_selector('div.page-nav > a')[j].click()
        sleep(1.5)
        for k in range(0, 50) : 
            albumBtn = driver.find_elements_by_css_selector('td.info > a.albumtitle')
            songTtl = driver.find_elements_by_css_selector('tr.list > td.info > a.title')
            chartdate.append(dates[i])
            titles.append(songTtl[k].text)
            ranks.append((j*50)+k+1)
            sleep(1)
            albumBtn[k].click()
            sleep(1.5)
            albumTtl = driver.find_element_by_css_selector('div.info-zone > h2.name').text
            albumTag = driver.find_elements_by_css_selector('ul.info-data > li >span.value')
            albums.append(albumTtl)
            artists.append(albumTag[0].text)
            genres.append(albumTag[1].text)
            publishers.append(albumTag[2].text)
            agencies.append(albumTag[3].text)
            pub_dates.append(albumTag[4].text)
            driver.back()
            sleep(1.5)
    globals()['top200_{}'.format(str(dates[i])[:6])] = pd.DataFrame({'chart': chartdate,
                                                                  'title' : titles,
                                                                  'rank' : ranks,
                                                                  'album' : albums,
                                                                  'artist' : artists,
                                                                  'genre' : genres,
                                                                  'publisher' : publishers,
                                                                  'agency' : agencies,
                                                                  'date' : pub_dates})
    print('top200_{} done'.format(str(dates[i])[:6]))
```

```python
temp = top200_202108.groupby(top200_202108['publisher']).count()
temp.index.values[12] = '지니뮤직, Stone Music Ent.' # 이름이 너무 길어서 수정
temp2 = temp.sort_values('album', ascending=False)
```

```python
print()
col_palette = sns.color_palette("tab20", 14) ## 색상
labels = temp2.index ## 라벨
values = temp2['album'] ## 빈도
 
fig = plt.figure(figsize=(20,20)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 배경색을 하얀색으로 설정
ax = fig.add_subplot() ## 프레임 생성

plt.title('2021년 8월 Genie Music Top 200 : 발매사별 M/S', fontsize=24)

pie = ax.pie(values, ## 파이차트 출력
       startangle=120, ## 시작점을 90도(degree)로 지정
       counterclock=False, ## 시계방향으로 그려짐
       colors = col_palette, ## 색상 지정
       )
 
total = np.sum(values) ## 빈도수 합
 
threshold = 5 ## 상한선 비율
sum_pct = 0 ## 퍼센티지
 
bbox_props = dict(boxstyle='square',fc='w',ec='w',alpha=0) ## annotation 박스 스타일
 
## annotation 설정
config = dict(arrowprops=dict(arrowstyle='-'),bbox=bbox_props,va='center')
 
for i,l in enumerate(labels):
    ang1, ang2 = ax.patches[i].theta1, ax.patches[i].theta2 ## 파이의 시작 각도와 끝 각도
    center, r = ax.patches[i].center, ax.patches[i].r ## 원의 중심 좌표와 반지름길이
    
    if i < len(labels) - 1:
        sum_pct += float(f'{values[i]/total*100:.1f}')
        text = f'{values[i]/total*100:.1f}%'
    else: ## 마지막 파이 조각은 퍼센티지의 합이 100이 되도록 비율을 조절
        text = f'{100-sum_pct:.1f}%'
    
    ## 비율 상한선보다 작은 것들은 Annotation으로 만든다.
    if values[i]/total*100 < threshold:
        ang = (ang1+ang2)/2 ## 중심각
        x = np.cos(np.deg2rad(ang)) ## Annotation의 끝점에 해당하는 x좌표
        y = np.sin(np.deg2rad(ang)) ## Annotation의 끝점에 해당하는 y좌표
        
        ## x좌표가 양수이면 즉 y축을 중심으로 오른쪽에 있으면 왼쪽 정렬
        ## x좌표가 음수이면 즉 y축을 중심으로 왼쪽에 있으면 오른쪽 정렬
        horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
        connectionstyle = "angle,angleA=0,angleB={}".format(ang) ## 시작점과 끝점 연결 스타일
        config["arrowprops"].update({"connectionstyle": connectionstyle}) ## 
        ax.annotate(text, xy=(x, y), xytext=(1.5*x, 1.2*y),
                    horizontalalignment=horizontalalignment, **config, fontsize=15)
    else:
        x = (r/2)*np.cos(np.pi/180*((ang1+ang2)/2)) + center[0] ## 텍스트 x좌표
        y = (r/2)*np.sin(np.pi/180*((ang1+ang2)/2)) + center[0] ## 텍스트 y좌표
        ax.text(x,y,text,ha='center',va='center',fontsize=24, color='white')
    
plt.legend(pie[0],labels,loc='upper right', fontsize=12) ## 범례
plt.show()
```
